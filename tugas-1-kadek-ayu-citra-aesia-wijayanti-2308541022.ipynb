```python
def user_choice(pilihan1,pilihan2):
    valid = False
    
    while(not valid):
        print("[",pilihan1,"/",pilihan2,"] : ", end = "")
        user_input_22 = input().lower()
        if(user_input_22 == pilihan1 or user_input_22 == pilihan2):
            valid = True
        else:
            print("Only answer the choice!")
            
    return user_input_22

def citra_bingung_22():
    print("Kemarin, kamu pp atau engga?")
    citra_22 = user_choice("pp","engga")
    
    if citra_22 == "pp":
        print("Nggak capek?")
        capek = user_choice("capek dong", "nggak si")
        
        if capek == "capek dong":
            print("istirahat yuk")
        else:
            print("semangat ya")
    else:
        print("tidak cape")

if __name__ == "__main__":
    citra_bingung_22();
```
